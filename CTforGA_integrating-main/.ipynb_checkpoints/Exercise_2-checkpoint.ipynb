```python
import pandas as pd

link3="https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
data3=pd.read_html(link3,header=0,flavor="bs4",attrs={'class':"wikitable sortable"})
```

```python
regiondata=data3[0].copy()
regiondata.head()
```

```python
WhichToDrop=list(set(range(8))-set([1,2]))
WhichToDrop
```

```python
regiondata.drop(labels=regiondata.columns[WhichToDrop],axis=1,inplace=True)
regiondata
```

```python
regiondata.Region.value_counts()
```

```python
regiondata=regiondata[regiondata.Region!="World"]
regiondata.reset_index(drop=True,inplace=True)
regiondata
```

```python
regiondata.columns=["Country","Region"]
```

```python
regiondata[regiondata["Country"].str.endswith("]")]
```

```python
regiondata.loc[:,'Country']=regiondata.Country.str.replace("\\[.+?\\]","",regex=True)
```

```python
sorted(regiondata["Country"].to_list())
```

```python
regiondata[['Country', 'CountryMain']] = regiondata.Country.str.split(pat='(', n=1, expand=True)
regiondata
```

```python
regiondata.drop("CountryMain",axis=1,inplace=True)
```

```python
regiondata
```

```python
regiondata=regiondata.apply(lambda x: x.str.strip())
```

```python
regiondata.to_csv("regiondata.csv",index=False)
```

```python
allDataFull=pd.read_csv("https://github.com/EvansDataScience/CTforGA_integrating/raw/main/allData.csv")
allDataFull
```

```python
leftCountry=set(allDataFull.Country)
len(leftCountry)
```

```python
rightCountry=set(regiondata.Country)
```

```python
len(leftCountry.intersection(rightCountry))
```

```python
UnmatchedLeft=list(leftCountry.difference(rightCountry))
UnmatchedLeft
```

```python
UnmatchedRight=list(rightCountry.difference(leftCountry))
UnmatchedRight
```

```python
from thefuzz import process
[(process.extract(left, UnmatchedRight,limit=2),left) for left in UnmatchedLeft]
```

```python
changesRight={'DR Congo':'Democratic Republic of the Congo',
'Congo':'Republic of the Congo',
'East Timor':'Timor-Leste'}


regiondata.Country.replace(changesRight,inplace=True)
```
